```python
# Delete short runs. Adapted from https://github.com/tensorflow/tensorboard/issues/785#issuecomment-571749832

from argparse import ArgumentParser
import shutil
from tensorboard.backend.event_processing.event_file_inspector import (
    get_inspection_units,
    print_dict,
    get_dict_to_print,
)

"""
Deletes all folders with small tensorboard run files
"""

run_len = {}
inspect_units = get_inspection_units(logdir="models")


# print(inspect_units)
for run in inspect_units:
    path = run[0]
    max_length = 0
    for key, value in get_dict_to_print(run.field_to_obs).items():
        if value is not None:
            length = value["max_step"]
            if max_length < length:
                max_length = length
    run_len[path] = max_length

to_archive = { k:v for k,v in run_len.items() if v < 1000 }

print(f"Total runs: {len(run_len)}")
print(f"To archive: {len(to_archive)}")
```

```python
import os
# Create the archive folder if it doesn't exist
if not os.path.exists("logs/archive"):
    os.makedirs("logs/archive")

for run in to_archive:
  out_folder = f"logs/archive/{run.replace('models/', '')}"
  # Move the run logs to logs/archive
  print(f"Moving {run} to logs/archive/{run}")
  shutil.move(run, f"logs/archive/{run}")
  
# for run, length in run_len.items():
#     if length < args.min_run_len:
#         if args.list:
#             print(f'{run} is {length} steps long and so will be deleted')
#         else:
#             try:
#                 print(f'{run} is {length} and was deleted')
#                 shutil.rmtree(run)
#             except:
#                 print(f"OS didn't let us delete {run}")
#     else:
#         print(f'{run} is {length} and is good')
```
